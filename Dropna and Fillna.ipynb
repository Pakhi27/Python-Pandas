```python
import pandas as pd
import numpy as np

# Example DataFrame with NaN values
data = {
    "a": [2.0, 3.0, np.nan, 5.0, 6.0],
    "b": [6.0, 7.0, np.nan, 9.0, 10.0],
    "c": [11.0, 12.0, np.nan, 14.0, 15.0],
    "d": [16.0, 17.0, np.nan, 19.0, 20.0],
    "e": [21.0, 22.0, np.nan, 24.0, 25.0]
}

var1 = pd.DataFrame(data)
print("Original DataFrame with NaN values:")
print(var1)

# Fill NaN values with specified values and reassign it to var1
var1 = var1.fillna({"a": "python", "b": "C", "c": "C++", "d": "java", "e": "sql"})
print("\nDataFrame after filling NaN values with specified values:")
print(var1)

# Alternatively, use inplace=True
# var1 = pd.DataFrame(data)  # Resetting var1 to original DataFrame for demonstration
# var1.fillna({"a": "python", "b": "C", "c": "C++", "d": "java", "e": "sql"}, inplace=True)
# print("\nDataFrame after filling NaN values with specified values (inplace=True):")
# print(var1)
var1.fillna(method="ffill")
```

```python
import pandas as pd
import numpy as np

# Example DataFrame with NaN values
data = {
    "a": [2.0, 3.0, np.nan, 5.0, 6.0],
    "b": [6.0, 7.0, np.nan, 9.0, 10.0],
    "c": [11.0, 12.0, np.nan, 14.0, 15.0],
    "d": [16.0, 17.0, np.nan, 19.0, 20.0],
    "e": [21.0, 22.0, np.nan, 24.0, 25.0]
}

var1 = pd.DataFrame(data)
print("Original DataFrame with NaN values:")
print(var1)
var1.fillna(method="bfill")
var1.fillna(method="bfill",axis=1)
var1.fillna(1,inplace=True)
```
