# reshaping panda file

# Melt

```python
import pandas as pd
var=pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var
```

```python
pd.melt(var)
#  id=first column or front indexing
pd.melt(var,id_vars=["eng"])
pd.melt(var,id_vars=["days"])
pd.melt(var,id_vars=["days"],var_name="python",value_name="wscube")
```

# Pivot()

```python
import pandas as pd
var=pd.DataFrame({"days":[1,1,1,1,2,2],"st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var
```

```python
# argument important
var.pivot(index="days",columns="st_name",values="eng")
```

```python
#  agg functions=mean or sum 
var.pivot_table(index="st_name",columns="days",aggfunc="mean",margins="True")
```
