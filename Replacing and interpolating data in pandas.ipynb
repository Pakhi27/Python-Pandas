# Replacing data in csv file

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Singh\\MY_CSV1.csv")
var
```

```python
# var=pd.DataFrame(dic)
var.replace(to_replace=1,value=11,inplace=True)
var
# var.replace(to_replace="python",value="C",inplace=True)
# var
```

```python
var.replace(to_replace=[1,2,3,4],value=11,inplace=True)
var
```

```python
var.replace("[A-Za-z]","python",regex=True)
var
```

```python
var.replace({"a":"[A-Za-z]"},22,regex=True)
```

```python
var.replace(1,method="ffill")
var.replace(1,method="bfill")
```

```python
var.replace(1,method="ffill",limit=2)
```

```python
#  replaces original data
var.replace(1,method="ffill",limit=2,inplace=True)
```

# Interpolate

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Singh\\MY_CSV1.csv")
var
```

```python
#  automatic fills the data according to gap only for digits not alphabets
var.interpolate()
```

```python
var.interpolate(method="linear")
```

```python
# col
var.interpolate(method="linear",axis=0)
# row data type should be same of all elements in the row
var.interpolate(method="linear",axis=1)
```

```python
#  fills only 1st 2 or 3 data according to limit
# var.interpolate(limit=2)
#  backward or forward or both
var.interpolate(limit_direction="forward",limit=2)
var
```

```python
#  intersection does not show any null values
var.interpolate(limit_area="inside")
#  union showa null values
var.interpolate(limit_area="outside")
#  inplace doesnt creates any copy
```
