# Join and append in data frames

```python
#  join
import pandas as pd
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var2=pd.DataFrame({"C":[10,20,30],"B":[11,22,33]},index=["a","b","c"])
var1.join(var2)
#  only min values are shown but using how we can print all the values
# var2.join(var1,how="right")
#  union null values are also shown
# var2.join(var1,how="outer",lsuffix="_12")
#  intersection only common values are shown
# suffix used to rename the columns
# var2.join(var1,how="inner",lsuffix="_12",rsuffix="_123")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var2=pd.DataFrame({"C":[10,20,30,40],"B":[11,22,33,44]},index=["a","b","c","d"])
var1.join(var2)
```

# Append

```python
import pandas as pd
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var2=pd.DataFrame({"C":[10,20,30],"B":[11,22,33]},index=["a","b","c"])
var1.append(var2)
#  append does not exist in latest version of pandas
var1.append(var2,ignore_index=True)
```
